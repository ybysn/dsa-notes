```python
def is_even(k):
    return k % 2 == 0

print(is_even(10))
print(is_even(7))
```

```python
def find_max(data):
    if len(data) == 0:
        raise ValueError('empty sequence')

    biggest = data[0]
    for value in data:
        if value > biggest:
            biggest = value
    return biggest

print(find_max([3, 1, 7, 2]))
print(find_max([-5, -1, -9]))
```

```python
a = [1, 2, 3]
b = a
b.append(4)

print(a)
print(b)
print(a is b)
```

```python
a = [1, 2, 3]
b = list(a)
b.append(4)

print(a)
print(b)
print(a is b)
```

```python
def square(n):
    return n * n

print(square(5))
```

```python
def is_odd(k):
    return k % 2 != 0
```

```python
def count_positive(data):
    count = 0
    for num in data:
        if num > 0:
            count += 1
    return count
```

```python
def sum_list(data):
    total = 0
    for num in data:
        total += num
    return total
```

```python
def contains(data, target):
    for num in data:
        if num == target:
            return True
    return False
```
